# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
```

```python
#數學函數表示方式
x = np.arange(-10,10,0.01)
y=list(map(lambda x:x if x>0 else 0,x))

#顯現ReLU函数
plt.plot(x,y)
plt.grid(color='b' , linewidth='0.1' ,linestyle='--')


# 顯現圖示的Title
plt.title('ReLU')

# 顯現 the Sigmoid formula
plt.text(-4, 8, r'$ReLU(x)= \max(0, x)$', fontsize=15)

# create the graph
plt.show()
```

```python
#顯現ReLU函数一階導數
dev_y =list(map(lambda x: 1 if x>0 else 0,x))
plt.plot(x,dev_y)
plt.grid(color='b' , linewidth='0.1' ,linestyle='--')

# 顯現圖示的Title
plt.title('ReLU')

# 顯現 the Sigmoid formula
plt.text(-10, 0.8,  r'$\frac{d}{dx} ReLU(x)= \mathcal{I}_{\{x>0\}}(x)$', fontsize=10)

# create the graph
plt.show()
```

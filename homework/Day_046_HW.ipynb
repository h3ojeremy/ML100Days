```python
from sklearn import datasets, metrics
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
digits = datasets.load_digits()
# 切分訓練集/測試集
x_train, x_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.25, random_state=4)
```

```python
#畫幾個 train data 給大家看
some_data = [5, 6, 7, 8, 9, 10]

for i in some_data:
    plt.figure()
    img_tmp = np.reshape(x_train[i,:], (8, 8))
    img = plt.imshow(img_tmp)
    img.set_cmap('gray')
    plt.axis('off')
    plt.title('label = %s'%str(y_train[i]))
```

```python
# 建立模型
clf = GradientBoostingClassifier()

# 訓練模型
clf.fit(x_train, y_train)

# 預測測試集
y_pred = clf.predict(x_test)
```

```python
#畫幾組預測結果出來
some_data = [5, 6, 7, 8, 9, 10]

for i in some_data:
    plt.figure()
    img_tmp = np.reshape(x_test[i,:], (8, 8))
    img = plt.imshow(img_tmp)
    img.set_cmap('gray')
    plt.axis('off')
    plt.title('our model predict label = %s, true label = %s'%(str(y_pred[i]), str(y_test[i])))
```

```python
#預測與實際不同的 diff_num 個例子
diff_num = 8
err_ind = y_test != y_pred #預測錯誤的 boolean array
all_index = np.array([i for i in range(len(y_test))])
all_diff_sample = all_index[err_ind];

if diff_num < len(all_diff_sample):
    all_diff_sample = all_diff_sample[:diff_num]

for idx, item in enumerate(all_diff_sample):
    plt.figure(num=idx)
    img_tmp = np.reshape(x_test[item,:], (8, 8))
    img = plt.imshow(img_tmp)
    img.set_cmap('gray')
    plt.axis('off')
    plt.title('our model predict label = %s, true label = %s'%(str(y_pred[item]), str(y_test[item])))
```
